```python
#Q2.
fmri = sns.load_dataset('fmri')
```

```python
#Q6.

flights = sns.load_dataset('flights')
```

```python
flights
```

```python
pivot_data = flights.pivot_table(index = 'month' , columns = 'year' , values = 'passengers')
plt.figure(figsize = (15,8))
sns.heatmap(pivot_data , cmap = 'coolwarm' , annot=True, linewidths=1)

plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Traffic Heatmap')
plt.show()
```
